```python
# 3.목표 매출과 비교해서 총 매출 수치는 어떠한가요?
months = ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb']
total_sales = [5280000, 5501000, 5469000, 5480000, 5533000, 5554000]
target_sales = [5280000, 5500000, 5729000, 5968000, 6217000, 6476000]

# 바플롯 그리기
plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
plt.bar(months, total_sales, label='Total Sales', color='blue', alpha=0.7)
plt.bar(months, target_sales, label='Target Sales', color='orange', alpha=0.7)
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.title('Total Sales vs Target Sales (Bar Plot)')
plt.legend()
plt.grid(True)

# 일반 플롯 그리기
plt.subplot(1, 2, 2)
plt.plot(months, total_sales, marker='o', label='Total Sales', color='blue')
plt.plot(months, target_sales, marker='o', label='Target Sales', color='orange')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.title('Total Sales vs Target Sales (Line Plot)')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()

print("매달 매출을 보았을 때 10월 까지 순 이익을 남겼고 11월 부턴 총 매출이 감소하여 목표매출 보다 떨어지고 있습니다")
```

```python
# 4.단가는 어떻게 되고있죠? 왜 계속 내려가고 있는 걸까요
months = ['Sep', 'Oct', 'Nov', 'Dec', 'Jan', 'Feb']
unit_price = [2.00, 2.00, 2.00, 1.90, 1.90, 1.90]

# 1온스별 단가 시각화
plt.figure(figsize=(10, 6))

plt.plot(months, unit_price, marker='o', color='green')
plt.xlabel('Month')
plt.ylabel('Unit Price ($)')
plt.title('1 Ounce Unit Price Trend')
plt.grid(True)

plt.show()

print("11월 까지 2$였던 온스당 단가가 12월 부터 1.9$로 내려갔습니다. \n1온스별 단가가 감소하는 이유는 다양할 수 있습니다. 예를 들면, 생산 비용의 변화, 시장 수요 및 공급 상황의 변동, 경쟁 업체의 가격 조정 등이 영향을 미칠 수 있습니다. ")
```
